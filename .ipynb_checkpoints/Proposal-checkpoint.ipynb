# INTRODUCTION

Mothers play a central role in nurturing and supporting their children's academic development, making their occupational status a significant factor in shaping educational trajectories. Exploring this correlation offers valuable insights into several key areas. Firstly, it can illuminate the impact of maternal employment on children's academic achievements. Mothers who are employed may have varying degrees of availability and involvement in their children's education, depending on factors such as work hours, flexibility, and job satisfaction. Understanding how different aspects of maternal employment relate to academic performance can help identify opportunities to support working mothers in balancing their professional and parental responsibilities effectively.

We aim to answer the question "can you predict mothers' occupations based on study time and final grade or absences and failures?".
Our model will use final grades and absences as predictors and our target variable will be the mother's job. 

The dataset from the UCI Machine Learning Repository, focusing on high school students' performance, provides a rich source of information that can facilitate the study of the correlation between a mother's occupation and their child's academic performance. # Preliminary exploratory data analysis:

First, we must load in required libraries.

```r
# Load necessary libraries
library(tidyverse)
library(readr)
library(repr)
library(tidymodels)
options(repr.matrix.max.rows = 6)
```

```r
#setting seed for all randomized processes to be reproducible
set.seed(24)
```

In the code block below, we are reading in the data set from the web. Because it is a zip file, it has to be unzipped.

```r
# Setting URL to read in
url <- "https://archive.ics.uci.edu/ml/machine-learning-databases/00320/student.zip"

# Temporary file path
temp <- tempfile()

# # Downloading the zip file
download.file(url, temp, mode = "wb")

# Unzipping the file
unzip(temp, exdir = "data")

# Read the CSV file directly from the extracted directory
student_data <- read_delim("data/student-por.csv", delim = ";")

# View the first few rows of the dataset
head(student_data)
```

The data is already tidy. Now, we will select the relevant columns and factorize our target variable.

```r
student_data_tidy <- student_data |>
    select(Mjob, failures, absences, G3) |>
    rename(final_grade = G3, mother_job = Mjob) |>
    mutate(mother_job = as_factor(mother_job))
    
    
head(student_data_tidy)
```

Now we will split our data into a training set and testing set.

```r
student_data_split <- initial_split(student_data_tidy, prop = 0.75, strata = mother_job)
student_data_train <- training(student_data_split)
student_data_test <- testing(student_data_split)
```

Now we will summarize the training set.

```r
student_data_summary <- student_data_train |>
    group_by(mother_job) |>
    summarize(count = n(),
            mean_absences = mean(absences),
            max_absences = max(absences),
            min_absences = min(absences),
            mean_final_grade = mean(final_grade),
            max_final_grade = max(final_grade),
            min_final_grade = min(final_grade),
            mean_failures = mean(failures),
            max_failures = max(failures),
            min_failures = min(failures))

head(student_data_summary)
```

There is class imbalance, but we will fix it by upscaling health and teacher. Since the ranges of the predictor variables are different, we should standardize the data by scaling and centering.
The max failures for each occupation is 3, except for teacher. The max number of absences are vastly different compared to the others.

Now we will plot our 2 models.


```r
options(repr.plot.width=10, repr.plot.height=7)
```

```r

student_plot_1 <- student_data_train |>
    ggplot(aes(x = final_grade, fill = mother_job)) +
    geom_histogram(binwidth = 2) +
    labs(x = "Final Grade", title = "Distribution of Students' Final Grade", fill = "Mother's Job") +
    theme(text = element_text(size = 20))

student_plot_1

student_plot_2 <- student_data_train |>
    ggplot(aes(x = absences, y = final_grade, color = mother_job)) +
    geom_point() +
    labs(x = "Absences", y = "Final Grade", title = "Absences v.s Final Grade", colour = "Mother's Job") +
    theme(text = element_text(size = 20))
    

student_plot_2
```
